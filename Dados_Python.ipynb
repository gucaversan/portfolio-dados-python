```python
mapeamento_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=mapeamento_colunas)
df.columns
```

```python
df["senioridade"].value_counts()
```

```python
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
mapeamento_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelance'
}

df['contrato'] = df['contrato'].replace(mapeamento_contrato)
df['contrato'].value_counts()
```

```python
mapeamento_tamanho_empresa = {
    'M': 'Media',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(mapeamento_tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapeamento_remoto = {
    0: 'Presencial',
    50: 'Hibrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(mapeamento_remoto)
df['remoto'].value_counts()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["ano"].unique()
```

```python
df[df.isnull().any(axis=1)]

# "isnull", pega todos os dados nulos. O metodo "any" verifica se tem pelo
# menos um valor nulo e o "axis=1", faz verificar linha por linha.

# Essa linha de código filtra e exibe todas as linhas do seu DataFrame que
# contêm pelo menos um valor nulo (missing value) em qualquer uma de suas
# colunas
```

```python
import numpy as np

# Inicialmente, a biblioteca NumPy é importada para nos permitir usar np.nan,
# que é a forma padrão de representar um valor "Não é um Número" ou um dado
# ausente em ambientes numéricos como DataFrames do Pandas.

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000],
})

# Um DataFrame simples é criado. Ele contém duas colunas: nome (com nomes de pessoas) e salario (com seus
# respectivos salários). Intencionalmente, alguns salários foram definidos como np.nan para simular dados
# ausentes.

df_salarios['salarios_media'] = df_salarios["salario"].fillna(df_salarios['salario'].mean().round(2))

# Esta é a parte central. O código calcula a média dos salários existentes na coluna salario
# (df_salarios['salario'].mean()) e arredonda esse valor para duas casas decimais (.round(2)).

df_salarios['salarios_mediana'] = df_salarios["salario"].fillna(df_salarios['salario'].median())

# Calcula mediana e substitui os nulos pela mediana.

df_salarios

# Propósito: O objetivo é preencher inteligentemente os dados ausentes de salário, usando a média dos salários
# disponíveis, o que é uma técnica comum de tratamento de dados para garantir que todas as entradas tenham um
# valor, mesmo que estimado.
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

# ffill = preenche com o valor anterior.

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

# bfill = preenche com o valor posterior.

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Salvador']
})

df_cidades['cidade_corrigida'] = df_cidades['cidade'].fillna('Não informado')

# O fillna() está sendo usado com o argumento 'Não informado', o que significa
# que todos os valores np.nan encontrados na coluna 'cidade' serão
# substituídos pela string "Não informado".

df_cidades
```
